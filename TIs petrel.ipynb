```python
import numpy as np
import os
import glob

def read_gslib_to_numpy(filepath):
    with open(filepath, 'r') as file:
        # Read the title (first line)
        title = file.readline().strip()
        
        # Read the second line: number of variables and names
        num_vars = int(file.readline().strip())
        var_names = [file.readline().strip() for _ in range(num_vars)]
        
        # Read the rest as data
        data = np.loadtxt(file)
        
    return data, var_names, title

# Directory containing the files
input_directory = "."  # Replace with your actual directory path
output_directory = "Define location"  # Replace with your desired output directory

# Get all files matching the pattern
file_pattern = os.path.join(input_directory, "*_frac_*_width_*_orient")
all_files = glob.glob(file_pattern)

# Process each file
for file_path in all_files:
    try:
        # Extract the base filename without path
        base_name = os.path.basename(file_path)
        print(f"Processing {base_name}...")
        
        # Read the data
        data, variable_names, title = read_gslib_to_numpy(file_path)
        
        # Reshape the data (assuming all files have the same dimensions)
        A = data.reshape(128, 128, 1000)
        
        # Save as numpy file with the same base name
        output_file = os.path.join(output_directory, f"{base_name}.npy")
        np.save(output_file, A)
        
        print(f"Saved to {output_file}")
        
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

print("All files processed!")
```
